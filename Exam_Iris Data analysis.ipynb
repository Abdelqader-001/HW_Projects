```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statistics as st
```

```python
%matplotlib inline
```

```python
df_iris=pd.read_csv('iris-data.csv')
```

```python
df_iris.head()    #  Display first 5 records
```

```python
df_versicolor=df_iris[df_iris.iris_class == 'Iris-versicolor']    # taking sample from  iris data fram for Iris-versicolor class
```

```python
df_versicolor
```

// Creating an array of (petal_length_cm) derived from df_versicolor data fram 

```python
versicolor_petal_length_arr = np.array(df_versicolor['petal_length_cm']) 
```

// Creating an array of (petal_width_cm) derived from df_versicolor data fram 

```python
versicolor_petal_Width_arr = np.array(df_versicolor['petal_width_cm'])
```

```python
st.pvariance(versicolor_petal_length_arr)  # calculating varince for petal_length
```

```python
st.pstdev(versicolor_petal_length_arr)    # calculating standard deviation for petal_length
```

```python
st.pvariance(versicolor_petal_Width_arr) #  calculating varince for petal_width
```

```python
st.pstdev(versicolor_petal_Width_arr)   # calculating standard deviation for petal_width
```

```python
versicolor_plot=plt.scatter(versicolor_petal_length_arr , versicolor_petal_Width_arr)  # scatter plot for versicolor_petal length and width
plt.title("versicolor petal length VS width")
plt.ylabel('petal_Width')
plt.xlabel('petal_Length')
```

```python
df_setosa=df_iris[df_iris.iris_class == 'Iris-setosa']
setosa_petal_length_arr = np.array(df_setosa['petal_length_cm']) 
setosa_petal_width_arr = np.array(df_setosa['petal_width_cm']) 
```

```python
df_virginica=df_iris[df_iris.iris_class == 'Iris-virginica']
virginica_petal_length_arr = np.array(df_virginica['petal_length_cm']) 
virginica_petal_width_arr = np.array(df_virginica['petal_width_cm']) 
```

```python
setosa_plot=plt.scatter(setosa_petal_length_arr , setosa_petal_width_arr)  # scatter plot for setosa_petal length and width
```

```python
virginica_plot=plt.scatter(virginica_petal_length_arr , virginica_petal_width_arr)  # scatter plot for virginica_petal length and width
```

```python
## scatter plot for All types petal length and width
virginica_plot=plt.scatter(virginica_petal_length_arr , virginica_petal_width_arr,color='red')
setosa_plot=plt.scatter(setosa_petal_length_arr , setosa_petal_width_arr,color='blue')
versicolor_plot=plt.scatter(versicolor_petal_length_arr , versicolor_petal_Width_arr,color='green')
plt.legend((virginica_plot,setosa_plot,versicolor_plot),('virginica plot','setosa plot','versicolor plot'))
plt.show()
```

the above scatter plot shows that the relation for all classes are incresing and the range of each class(i.e. the versicolor range is between 3cm - 5.5cm petal length), so we can conclude the class of Iris from it petal length

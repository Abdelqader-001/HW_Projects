```python
df_versicolor=df_iris[df_iris.iris_class == 'Iris-versicolor']    # taking sample from  iris data fram for Iris-versicolor class
```

```python
df_versicolor
```

// Creating an array of (petal_length_cm) derived from df_versicolor data fram 

```python
versicolor_petal_length_arr = np.array(df_versicolor['petal_length_cm']) 
```

// Creating an array of (petal_width_cm) derived from df_versicolor data fram 

```python
versicolor_petal_Width_arr = np.array(df_versicolor['petal_width_cm'])
```

```python
st.pvariance(versicolor_petal_length_arr)  # calculating varince for petal_length
```

```python
st.pstdev(versicolor_petal_length_arr)    # calculating standard deviation for petal_length
```

```python
st.pvariance(versicolor_petal_Width_arr) #  calculating varince for petal_width
```

```python
st.pstdev(versicolor_petal_Width_arr)   # calculating standard deviation for petal_width
```

```python
plt.scatter(versicolor_petal_length_arr , versicolor_petal_Width_arr )  # scatter plot for length and width
plt.title("versicolor petal length VS width")
plt.ylabel('petal_Width')
plt.xlabel('petal_Length')
```
